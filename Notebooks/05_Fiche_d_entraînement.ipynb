Maintenant que nous pouvons extraire la valeur RMSSD d'un ECG et que nous connaissons une interprétation de celui-ci, nous pouvons établir 
une méthode d'entraînement qui va respecter la periode de récupération minimum d'un athhèle pour optimiser son gain de performance. Ce temps de récupération peut être compléter d'un indicateur de charge d'entraînment et de test de performance régulier pour suivre l'avancer de l'athlète. 

## Quand s'entrainer ? 

La personnalisation d'un entraînement peut viser à déterminer le contenu de l’entraînement à venir (i.e. intensité). Cette personnalisation se base sur les résultats de l’enregistrement de variabilité cardiaque du jour.
- Si l’athlète observe une baisse de l’activité du système nerveux autonome comparativement à l’enregistrement de la veille →  il devra réaliser une séance d’entraînement de basse intensité puisque les auteurs (à cause des traces de fatigue )
- Si  l’enregistrement est supérieur à celui de la veille →  l’athlète devra réaliser une séance de haute intensité (car plus en forme que la veille).

- HRV- et HRV+  :  une diminution ou une augmentation de l’activité du système nerveux autonome du jour comparativement à la mesure de la veille

- Rest : jour de repos

**En français, cet algorithme nous dit que si on obseve une augmentation de la VFC (l'athèle est en forme) on augmente l'intensité de l'entraînement à suivre et inversement. Si on remarque plusieurs jours d'affilé une baisse de la VFC (l'athlète est fatigué ou en surentraînement) on lui propose donc de se reposer.**

## Ma séance est-elle adaptée à mon niveau ? 

Reprenons une expérience déjà réalisée, 11 athlètes d’un bon niveau (VMA = 18,9±1,2 km.h-1) ont effectué aléatoirement 4 séances d’entraînement avec, pour chacune, une intensité (70% - 100% VMA), une durée (41min – 69min) et un mode (continu - intermittent) différent. La charge d’entraînement a été évaluée à l’aide des méthodes classiques de Foster et Banister mais également par l’intermédiaire de 3 mesures RMSSD d’une durée de 5 minutes. La première étant réalisée juste avant l’entraînement (Pre5) et les deux autres étant effectuées respectivement immédiatement (Post5) et 30 minutes après l’exercice (Post30). Par la suite, la charge d’entraînement était calculée selon la formule suivante :

$$
TL_{HRV}= ln(T*\frac{Pre5 - Post5}{Post30 - Post5})
$$

Légende : 
- T: Durée de l’entraînement (min)

Les résultats ont montré que les 4 séances ont provoqué une diminution significative de l’activité du système nerveux autonome (Pre5>Post5). Les 30 minutes de récupération ayant permis, pour tous les exercices, de faire remonter significativement le RMSSD (Post5<Post30). La baisse de RMSSD induite par l’exercice était fortement corrélée à l’intensité de ce dernier et non à sa durée. 

La mesure Pre5 nous renseigne sur le niveau de fatigue actuel de l’athlète. La mesure Post5 caractérisé par une décroissance de RMSSD reflète l’intensité de l’exercice.

En résumé, la nouvelle formule permet de quantifier la charge d’un entraînement de manière totalement objective. Elle est difficilement utilisable dans le cadre de mesures quotidiennes (plannification de mesures), mais elle trouve un intérêt particulier dans le contexte du suivi des athlètes. En effet, il est tout à fait envisageable d’effectuer, à intervalles réguliers, une séance type dont la charge serait quantifiée à l’aide de la méthode. Ainsi, nous pourrions suivre l’adaptation à l’entraînement d’un athlète face à une séance étalon. Dans la logique la charge d'entraînment devrait diminuer à mesure de la progression de l'entraînement. Un autre test simple pour évaluer un entraînment serait de réaliser un test de VMA régulier pour suivre de même les progrès du coureur. Suivant les objectifs du coureur on peut imaginer une augmentation progressive de la difficulté du programme qui entraînerai un maintient de la charge d'entraînement perçu.
