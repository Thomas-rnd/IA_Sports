# Calcul RMSSD

# Théorie

Formule méthode RMSSD (Root Mean Square of the Successive Differences)
$$
RMSSD= \sqrt{\frac{1}{N}\sum_{i=1}^{N} (RR_{i}-RR_{i-1})^2}
$$

Une récente publication se focalisant sur le marqueur temporel RMSSD a montré une diminution du RMSSD lors de la phase de surentraînement d’un athlète et un retour aux valeurs de base à la suite d’une période de récupération.

Des scientifiques, Pichot et al. évaluent l’activité du système nerveux autonome durant 4 semaines à l’aide d’enregistrements de variabilité cardiaque bi-hebdomadaire réalisés la nuit. Les trois premières semaines sont composées de séances d’entraînement intensives et la dernière semaine d’un repos relatif. Comparativement à la première semaine, le RMSSD diminue en moyenne de 17 % puis 39 % lors des semaines 2 et 3 avant de remonter à 120 % lors de la dernière semaine. La hausse significative du RMSSD lors de la période d'affûtage corrélée à la diminution du niveau de fatigue des athlètes (i.e. baisse significative du score de fatigue) permet aux auteurs de conclure sur l’intérêt de l’utilisation de la variabilité cardiaque pour optimiser et individualiser les planifications d’entraînements des athlètes de façon totalement objective.

Une autre étude propose de déterminer le contenu de l’entraînement à venir (i.e. intensité) en fonction du résultat de l’enregistrement de  variabilité cardiaque du jour. En d’autres termes, si l’athlète observe une baisse de l’activité du système nerveux autonome comparativement à l’enregistrement de la veille, il devra réaliser une séance d’entraînement de basse intensité puisque les auteurs considèrent qu’il montre des traces de fatigue. A l’inverse, si l’enregistrement est supérieur à celui de la veille, l’athlète devra réaliser une séance de haute intensité puisque les auteurs considèrent qu’il est en plus grande forme que la veille (i.e. aptitude).

```python
from math import *

#Calcul RMSSD pics étant la liste de pics associés à un complexe QRS
def RMSSD(pics):
  res=0
  for i in range(len(pics)-1):
    res+=(pics[i+1]-pics[i])^2
  return sqrt(1/len(pics)*res)
  
```
